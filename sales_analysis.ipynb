```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
f= files.upload()
```

```python
data=np.genfromtxt('sales_dataset.csv',delimiter=',',skip_header=1,dtype=None,names=['Date','Category','Product','Sales','Advertisement'],encoding='utf-8') # contains unicode and encoding
print(data)
```

```python
#np.nan_to_num used to clean the invalid numeeric values (Nan - Not a Number)
sales_data=np.nan_to_num(data['Sales'])
advertisement_data=np.nan_to_num(data['Advertisement'])
```

```python
#normalizing data

sales_norm=(sales_data-np.mean(sales_data)/np.std(sales_data))
advertisement_norm=(advertisement_data-np.mean(advertisement_data)/np.std(advertisement_data))
```

```python
sales_mean=np.mean(sales_data)
sales_meadian=np.median(sales_data)
sales_std=np.std(sales_data)
print("Mean of sales=",sales_mean)
print("Median of sales=",sales_meadian)
print("Standard deviation of sales=",sales_std)
```

```python
# creating a variable to store unique productus

products=np.unique(data['Product'])
p=[]
s=[]

for product in products:
  sales=np.sum(data[(data['Product']==product)]['Sales'])
  p.append(product)
  s.append(sales)
print(p)
print(s)
```

```python
cats=np.unique(data['Category'])
pc=[]
cs=[]

for cat in cats:
  sales=np.sum(data[(data['Category']==cat)]['Sales'])
  pc.append(cat)
  cs.append(sales)
print(pc)
print(cs)
```

```python
# maximum value
max_s=np.max(s)
ind_s=s.index(max_s)   # index
print("Product with highest sales="+str(p[ind_s]))

max_cs=np.max(cs)
ind_cs=cs.index(max_cs)
print("Category with highest sales="+str(pc[ind_cs]))
```

```python
avg = np.average(data['Sales'])
print("Average Sales:"+str(avg))
```

```python
# to find avrage of specific priblems
avg_jeans= np.average(data[data['Product']=='Jeans']['Sales'])
print("Average of Jeans Sales:"+str(avg_jeans))
```
